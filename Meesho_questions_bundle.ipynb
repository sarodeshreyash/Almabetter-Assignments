### **2. What is Bayes’ Theorem? What is posterior and  likelihood?**

Bayes' Theorem is a statistical formula that helps to calculate the probability of an event, based on prior knowledge and new evidence. It is named after the English statistician Thomas Bayes.

In simple terms, Bayes' Theorem is used to update the probability of an event, based on new evidence that becomes available. It provides a way to combine prior knowledge (such as the probability of an event happening based on past experience) with new evidence (such as the results of a new experiment or observation) to arrive at a revised probability.

Bayes' Theorem is based on two key concepts: posterior probability and likelihood.

**Posterior probability** is the revised probability of an event, based on both prior knowledge and new evidence. It represents our updated belief about the probability of the event happening.


**Likelihood** is the probability of observing the new evidence, given that the event has occurred. It represents the strength of the evidence in favor of the event happening.


In Bayes' Theorem, the posterior probability is calculated by multiplying the prior probability by the likelihood, and then dividing by the marginal likelihood (the total probability of the evidence). 

Mathematically, this can be expressed as:

Posterior Probability = (Prior Probability * Likelihood) / Marginal Likelihood

### **3. What is dimensionality reduction? What is the difference between LDA and PCA?**

Dimensionality reduction is a technique used in machine learning and data analysis to reduce the number of variables in a dataset without losing too much information. It helps to simplify the dataset and make it more manageable for analysis, visualization, and modeling.

PCA (Principal Component Analysis) and LDA (Linear Discriminant Analysis) are two popular dimensionality reduction techniques used in data science.

**PCA (Principal Component Analysis)** is a technique used for dimensionality reduction in machine learning and data analysis. Its main goal is to identify the most important patterns and relationships in a dataset by reducing the number of dimensions, while retaining as much information as possible.

PCA works by transforming the data into a new coordinate system, where each dimension (or feature) is a linear combination of the original features. The first principal component is the direction in this new space that captures the most variation in the data. The second principal component is the direction that captures the most remaining variation, orthogonal to the first one. And so on, for the remaining components.

These principal components can be thought of as new axes that define the most important directions of the data. By projecting the data onto these new axes, we can represent the original data in a lower-dimensional space, while retaining as much of the original variance as possible.

The resulting lower-dimensional representation of the data can be used for further analysis, such as visualization or modeling. Additionally, the principal components themselves can provide insights into the most important patterns and relationships in the data.

**LDA (Linear Discriminant Analysis)** is a dimensionality reduction technique used in machine learning for supervised classification problems. Its main goal is to find a lower-dimensional representation of the data that maximizes the separation between different classes in the dataset.

LDA works by finding a set of linear combinations of the original features (called discriminant functions) that maximize the ratio of the between-class variance to the within-class variance. In other words, it tries to find directions in the data that separate the different classes as much as possible, while keeping the data points within each class as close together as possible.

The resulting discriminant functions can be used to project the data onto a lower-dimensional space, while preserving as much information as possible about the class labels. This lower-dimensional representation of the data can then be used for further analysis or modeling.

LDA assumes that the data is normally distributed and that the covariance matrices of each class are equal. It is also important to note that LDA is a supervised technique, meaning that it requires labeled data (i.e., data with known class labels) in order to work effectively.

### **8. What is Central Limit Theorem? What general conditions must be satisfied for the central limit theorem to hold?**

The Central Limit Theorem is a fundamental result in probability theory and statistics, which states that the sampling distribution of the mean of a large number of independent and identically distributed (i.i.d.) random variables will tend to follow a normal distribution, regardless of the underlying distribution of the individual random variables.

In simpler terms, the Central Limit Theorem says that if we take many samples of size "n" from any distribution, and calculate the mean of each sample, the distribution of those sample means will tend to be normal, as the sample size "n" gets larger and larger.

The general conditions that must be satisfied for the Central Limit Theorem to hold are:

Independence: The samples must be independent of each other. This means that the value of one sample should not influence the value of another sample.

Identical distribution: The samples must be drawn from the same underlying population, with the same mean and standard deviation.

Finite variance: The population distribution must have a finite variance. If the variance is infinite, the Central Limit Theorem may not hold.

Sufficiently large sample size: The sample size "n" should be large enough. A commonly used rule of thumb is that the sample size should be at least 30, although this may vary depending on the distribution of the individual random variables.

These conditions ensure that the sample means are normally distributed, which is useful for statistical inference and hypothesis testing. The Central Limit Theorem is widely used in various fields, such as finance, engineering, and social sciences, to estimate the mean and standard deviation of a population based on a sample.

### **9. What is N-gram Language Model? What is the difference between Unigram, Bigram and Trigram in NLP? How do N-gram Language Models work? Also mention its applications.**

An N-gram language model is a type of statistical language model that predicts the likelihood of a sequence of words based on the frequency of its constituent N-grams. An N-gram is a contiguous sequence of "N" items from a given sequence of text, usually words in the context of natural language processing.

In NLP, the most commonly used N-gram models are the unigram, bigram, and trigram models, which differ in the size of the N-gram. A unigram model considers each word as a separate unit, while a bigram model considers pairs of adjacent words, and a trigram model considers triples of adjacent words.

For example, consider the sentence "The cat sat on the mat." The unigram model would count the frequency of each individual word: "The" (1), "cat" (1), "sat" (1), "on" (1), "the" (1), and "mat" (1). The bigram model would count the frequency of each pair of adjacent words: "The cat" (1), "cat sat" (1), "sat on" (1), "on the" (1), "the mat" (1). The trigram model would count the frequency of each triple of adjacent words: "The cat sat" (1), "cat sat on" (1), "sat on the" (1), "on the mat" (1).

N-gram models work by estimating the probability of a particular sequence of words using the chain rule of probability:

P(w_1, w_2, ..., w_n) = P(w_1) * P(w_2 | w_1) * P(w_3 | w_1, w_2) * ... * P(w_n | w_1, w_2, ..., w_{n-1})

where w_1, w_2, ..., w_n are the words in the sequence.

The probability of each N-gram is estimated based on its frequency in a given corpus, and the probabilities of longer sequences are calculated by multiplying the probabilities of the constituent N-grams.

**N-gram models have many applications in NLP, including:**

Language modeling: predicting the next word in a sequence, given the preceding words.

Text generation: generating new text that follows the statistical patterns of a given corpus.

Speech recognition: transcribing spoken words into text.

Machine translation: translating text from one language to another.

Sentiment analysis: determining the sentiment (positive, negative, or neutral) of a given text.

However, N-gram models have limitations in capturing long-range dependencies and understanding the semantic meaning of text, and more advanced models such as neural language models have been developed to overcome these limitations.

### **10. Executing a binary classification tree algorithm is a simple task. But how does tree splitting take place? How does the tree determine which variable to break at the root node and which at its child nodes?**

Binary classification trees are constructed by recursively splitting the data into two groups based on a selected splitting criterion until a stopping criterion is met. The goal is to create a tree that maximally separates the classes in the data.

The process of tree splitting begins with selecting the best variable and its corresponding split point to split the data. The splitting criterion is chosen based on a metric that maximizes the separation of the two classes in the resulting subsets. One commonly used splitting criterion is the Gini index, which measures the impurity of a node based on the distribution of the class labels in that node. The variable with the lowest Gini index is chosen as the splitting variable.

Once the splitting variable is chosen, the data is partitioned into two subsets based on the selected split point. The split point is chosen based on the value that maximizes the splitting criterion. For continuous variables, the split point is usually chosen by sorting the values of the variable and selecting the mid-point between adjacent values. For categorical variables, each category is considered a separate subset.

The process of splitting is then repeated for each subset, continuing recursively until a stopping criterion is met, such as reaching a maximum tree depth, having too few samples in a node, or having too little information gain in a split.

The algorithm determines which variable to split at the root node and which at its child nodes based on the splitting criterion. At each level of the tree, the splitting criterion is computed for each variable, and the variable with the lowest criterion is chosen as the splitting variable. This process is repeated for each subsequent level until the stopping criterion is met.

In summary, the tree splitting process in binary classification trees involves selecting the best variable and split point based on a splitting criterion that maximizes the separation of the classes in the data. The algorithm determines which variable to split at the root node and which at its child nodes based on the splitting criterion, repeating the process recursively until a stopping criterion is met.

### **11. What are the different types of gradient descent algorithm? Differentiate  between batch, mini batch and stochastic gradient descent.**

Gradient descent is a popular optimization algorithm used in machine learning and deep learning. It is a first-order iterative optimization algorithm for finding the minimum of a cost function.

There are three main types of gradient descent algorithms: batch gradient descent, mini-batch gradient descent, and stochastic gradient descent.

Batch Gradient Descent:
Batch gradient descent is the most basic form of gradient descent. In batch gradient descent, the model updates its weights after processing the entire training set. This approach makes the algorithm computationally expensive but guarantees convergence to a global minimum.

Mini-Batch Gradient Descent:
Mini-batch gradient descent is a variation of batch gradient descent that updates the weights after processing a subset of the training set. This subset, called a mini-batch, is typically a small random sample of the training set. This approach reduces the computational cost of batch gradient descent while still achieving fast convergence.

Stochastic Gradient Descent:
Stochastic gradient descent is another variation of batch gradient descent that updates the weights after processing each individual training sample. This approach greatly reduces the computational cost of batch gradient descent but can also cause the algorithm to converge to a local minimum instead of the global minimum.

Here are the differences between the three types of gradient descent algorithms:

Batch gradient descent processes the entire training set, while mini-batch gradient descent processes a subset of the training set and stochastic gradient descent processes a single training example at a time.
Batch gradient descent guarantees convergence to a global minimum, while mini-batch and stochastic gradient descent can converge to a local minimum.
Batch gradient descent requires more computational resources compared to mini-batch and stochastic gradient descent because it processes the entire training set, while the other two only process a subset or individual examples at a time.
Mini-batch gradient descent can achieve a balance between convergence speed and computational resources by processing a small random sample of the training set at each iteration.
Stochastic gradient descent can be noisy and converge to a local minimum, but it is generally faster than batch and mini-batch gradient descent due to the frequent updates to the weights.

### **12. What is activation function in neural networks and what are its types? What is the difference between Sigmoid, Relu and Softmax activation fuction and which one is better?**

In a neural network, an activation function is a non-linear mathematical function that transforms an input signal into an output signal. The output signal is then used as input for the next layer of the neural network. Activation functions play a crucial role in the neural network's ability to learn and make accurate predictions.

There are several types of activation functions used in neural networks, but some of the most popular ones are:

Sigmoid Function:
The sigmoid function is a popular activation function that squashes the input values between 0 and 1. It is a smooth and continuous function, and its output can be interpreted as the probability of the input belonging to a particular class. However, the sigmoid function can suffer from the vanishing gradient problem, making it less effective in deep neural networks.

ReLU Function:
The ReLU (Rectified Linear Unit) function is a popular activation function that sets all negative values to zero and keeps all positive values as they are. It is a non-linear function that is easy to compute and helps overcome the vanishing gradient problem. The ReLU function is widely used in deep neural networks and has been shown to produce better results than other activation functions in many cases.

Softmax Function:
The softmax function is a popular activation function used in the output layer of a neural network for multi-class classification problems. It takes an input vector and squashes the values between 0 and 1 while maintaining their relative proportions. The output of the softmax function can be interpreted as the probabilities of the input belonging to each class.

The choice of activation function depends on the type of problem being solved and the structure of the neural network. The ReLU function is generally considered to be the default choice for most situations, as it is computationally efficient and has been shown to work well in deep neural networks. However, the sigmoid and softmax functions are more appropriate for specific tasks such as binary or multi-class classification, respectively.
