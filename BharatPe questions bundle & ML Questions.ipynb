### **2. Explain how  Decision Tree Algorithm works? What is pruning in decision trees and what are its types?**

The Decision Tree algorithm is a supervised machine learning algorithm that constructs a tree-like model to make decisions. The algorithm recursively **splits the dataset into subsets** based on the value of a selected feature until the subsets are as **homogeneous** as possible. This allows the algorithm to classify new data points by following the path from the root node to the leaf node that corresponds to the predicted class or group.

The algorithm builds a tree-like model of decisions and their possible consequences. It starts by selecting a feature from the dataset that best splits the data into homogeneous subgroups based on the value of that feature. This is done by measuring the information gain or entropy of each feature, which indicates the amount of uncertainty in the data that is reduced by splitting the data using that feature.

Once a feature is selected, the dataset is split into two or more subsets based on the value of that feature. The process is repeated on each subset until a stopping criterion is met, such as when all the data in a particular subset belongs to the same class or when a maximum depth of the tree is reached.

At each step of the algorithm, the decision tree model is built by selecting the feature that provides the best information gain or entropy reduction, and then using that feature to split the data into subsets. The process continues until a stopping criterion is met.

Once the decision tree is built, it can be used to classify new data by traversing the tree from the root to a leaf node based on the values of the input variables. The class label associated with the leaf node is then assigned to the input data


**Pruning** is a technique used to prevent overfitting in decision trees.
This is done by removing branches or nodes that do not contribute much to the classification accuracy. 

**Pre-pruning:** Pre-pruning involves stopping the tree construction early, before the tree becomes too complex and overfits the data. Pre-pruning techniques include:

Maximum Depth: Stopping the tree construction when a maximum depth of the tree is reached.

Minimum Samples per Leaf: Stopping the tree construction when a minimum number of samples is reached in a leaf node.

Minimum Impurity Decrease: Stopping the tree construction when the impurity decrease of a split is below a threshold.

Pre-pruning is computationally efficient and can save time and resources, but it might lead to underfitting if the stopping criterion is too strict.

**Post-pruning:**

Post-pruning involves removing branches or nodes from a fully grown tree, based on their estimated contribution to the overall accuracy of the tree.Post-pruning techniques include:

Reduced Error Pruning: Starting from the leaves of the tree, removing each node and testing the accuracy of the tree on the validation set. If the accuracy does not decrease significantly, the node is removed.

Cost Complexity Pruning: Introducing a trade-off between the accuracy and the complexity of the tree, by adding a penalty term to the impurity or the information gain measure. The penalty term is a function of the number of nodes or branches in the tree, and it controls the degree of pruning.

### **3. What is  Random Forest algorithm? How does random forest  reduce overfitting?**

The Random Forest algorithm is a popular machine learning algorithm that uses an ensemble learning approach to improve accuracy and reduce overfitting. 

The process of building a Random Forest model involves the following steps:

1. Randomly select a subset of the training data, with replacement (bootstrap sampling). This creates a new dataset that is the same size as the original dataset, but with some of the data points repeated and others left out.

2. Build a decision tree using the bootstrap sample. At each node of the tree, randomly select a subset of the features and choose the one that provides the best split.

3. Repeat steps 1 and 2 to build multiple decision trees. The number of trees in the forest is a hyperparameter that can be tuned to optimize the performance of the model.

4. For classification tasks, the predicted class for a new data point is determined by taking the majority vote of the classes predicted by each individual tree. For regression tasks, the predicted value is the average of the values predicted by each tree.
**Overfitting** occurs when a model is too complex and fits the noise in the training data rather than the underlying patterns. Random Forest reduces overfitting by using two key techniques: bootstrap sampling and feature randomness.

**Bootstrap sampling:** Random Forest creates multiple decision trees by randomly selecting a subset of the training data with replacement. This means that some data points may be selected multiple times and others may be left out. By creating different subsets of the data for each tree, Random Forest reduces the risk of overfitting to specific patterns in the data. The remaining data that are not selected for each tree can be used for testing the model's performance.

**Feature randomness:** At each node of the decision tree, Random Forest randomly selects a subset of the features to consider for the split. This means that different decision trees in the forest are built on different subsets of the features. By creating different subsets of the features for each tree, Random Forest reduces the risk of overfitting to specific features in the data. Additionally, it allows the algorithm to consider a more diverse set of features, which can help to capture more information about the data and improve the model's performance.

### **4. What is the difference between Precision and Recall? Explain with example** 

Precision and Recall are two crucial metrics used to evaluate the performance of a classification model in machine learning. 

Precision measures the ratio of true positive predictions to the total number of positive predictions made by the model. 
Precision can be calculated as:

precision = true positives / (true positives + false positives)

In contrast, Recall measures the ratio of true positive predictions to the total number of actual positive instances in the dataset.
Recall can be calculated as:

recall = true positives / (true positives + false negatives)

**Precision measures the accuracy of positive predictions, while recall measures the completeness of positive predictions**

### **5. What is KMeans?  How do you find optimal value of k  in KMeans algorithm?**

KMeans is a machine learning algorithm used to group similar data points together into clusters in an unsupervised manner. The algorithm works by assigning data points to the nearest cluster centroid and re-calculating the centroid of each cluster until the centroids no longer move significantly.

The algorithm works by:

1. Initialization:  KMeans begins by randomly initializing K centroids from the dataset. These centroids are used as the initial cluster centers. The algorithm requires the user to specify the number of clusters (K) beforehand.

2. Assigning data points to clusters: The algorithm then iteratively assigns each data point to the cluster whose centroid is nearest to it, based on the Euclidean distance between the data point and each centroid. The result is that each data point is assigned to a specific cluster.

3. Updating centroids: After assigning each data point to a cluster, the algorithm computes the mean of all data points in each cluster to determine the new centroid for that cluster.

4. Repeat until convergence: The algorithm repeats steps 2 and 3 until there is no change in the assignment of data points to clusters, which means that the algorithm has converged.

5. Output: The output of the KMeans algorithm is the final set of K clusters and their respective centroids. Each data point is assigned to one and only one cluster.

6. Optimization: KMeans is an optimization problem that seeks to minimize the sum of the squared distances between each data point and its assigned centroid. This objective function is also known as the "within-cluster sum of squares" (WCSS) or "inertia." The algorithm tries to minimize this function by adjusting the centroids and the assignment of data points to clusters.

7. Choosing the number of clusters: The number of clusters (K) is a hyperparameter that needs to be specified by the user. There are various methods to choose K, such as the elbow method, silhouette method, or gap statistic.

8. Limitations: KMeans is sensitive to the initial choice of centroids and may converge to a local minimum instead of the global minimum. It is also not suitable for datasets with non-spherical or overlapping clusters.

To determine the best value of k, the number of clusters, in KMeans algorithm, there are different methods to choose from. These methods include:

**Elbow method:** calculates the sum of squared distances between data points and their assigned cluster centroid for different values of k. The optimal value of k is the one where the rate of decrease in the sum of squared distances slows down and forms an elbow in the graph.

**Silhouette coefficient:** measures the similarity of data points to their assigned cluster compared to other clusters. The average silhouette coefficient is computed for different values of k, and the optimal value of k is the one that maximizes the average silhouette coefficient.

**Gap statistic:** compares the sum of squared distances for different values of k with that of a reference dataset generated by a random distribution. The optimal value of k is the one for which the gap between the two sums of squared distances is the largest.

**Hierarchical clustering:** involves building a hierarchical tree of clusters and using the number of branches in the tree to determine the optimal value of k.

### **8. What is the difference between tokenisation, lemmatization and stemming?**

Tokenization, stemming, and lemmatization are common techniques used in natural language processing to preprocess text data. 

**Tokenization** is the process of breaking down a piece of text into smaller units, called tokens, which usually correspond to words or sub-words. It is the first step in natural language processing (NLP) and allows for easier analysis of text data.

**Stemming** is the process of reducing inflected or derived words to their base or root form, known as a stem. This is done by removing suffixes or prefixes from words, so that words with the same stem are treated as identical. This can help to improve text search and analysis by reducing the number of unique words in a dataset.

**Lemmatization** is similar to stemming in that it also involves reducing words to their base form, but it does so using a more sophisticated approach that takes into account a word's context and part of speech. The resulting base form is called a lemma, and it can be used to group together related words that have the same root but different inflections.

### **11. Describe Hypothesis Testing. How is the statistical significance of an insight assessed? What is Alpha in hypothesis testing?**

Hypothesis testing is a statistical technique used to determine if a given hypothesis about a population is supported by sample data. It involves two hypotheses: a null hypothesis and an alternative hypothesis.

The null hypothesis is typically assumed to be true until there is sufficient evidence to reject it. The alternative hypothesis is usually what the researcher is interested in proving, and it is often formulated as a two-sided or one-sided test, depending on the nature of the research question.


Statistical significance of an insight is determined by evaluating the probability of obtaining the observed results by chance. This is done by calculating a p-value, which is the probability of obtaining a sample result as extreme or more extreme than the observed result, assuming that the null hypothesis is true. If the p-value is lower than a selected significance level (usually 0.05 or 0.01), the null hypothesis is rejected, and the alternative hypothesis is accepted.


Alpha, also known as the significance level, is the likelihood of rejecting the null hypothesis when it is actually true. It is a threshold for deciding whether to reject or fail to reject the null hypothesis. Typically, alpha is set to 0.05, which implies a 5% chance of rejecting the null hypothesis when it is actually true. A lower alpha level, such as 0.01, shows greater confidence in the results, but may also increase the likelihood of a type II error, which is failing to reject the null hypothesis when it is false.

### **12. What is  the difference between Label encoding and One Hot Encoding?**

**Label Encoding**: In Label Encoding, each category is assigned a unique integer number. This means that the categories are ordered and there is a sense of hierarchy among them. For example, if we have a categorical feature called "size" with categories "small", "medium", and "large", we could encode them as 0, 1, and 2 respectively. One potential issue with Label Encoding is that it may create a false sense of order or hierarchy when there is none, especially when using numerical algorithms. This can lead to incorrect results and bias in the model.

**One-Hot Encoding**: In One-Hot Encoding, each category is represented as a binary vector of 0s and 1s. The length of the vector is equal to the number of categories in the feature, and each category has a unique position in the vector. For example, if we have a categorical feature called "color" with categories "red", "green", and "blue", we could encode them as [1, 0, 0], [0, 1, 0], and [0, 0, 1] respectively. One-Hot Encoding is useful when there is no inherent order or hierarchy among the categories, and when the number of categories is relatively small. However, it can lead to a large number of features in the dataset, especially when there are many categories. This can lead to the "curse of dimensionality", where the model becomes inefficient and overfitting can occur.

Basically, The main difference between Label Encoding and One-Hot Encoding is the way they represent categorical data. Label Encoding assigns a unique integer to each category, which can be useful when the categories have some kind of order or hierarchy. One-Hot Encoding, on the other hand, represents each category as a binary vector, which can be useful when there is no order or hierarchy among the categories.

However, One-Hot Encoding can lead to a large number of features in the dataset, especially when there are many categories. In such cases, Label Encoding can be a more efficient way to represent the data. The choice of which encoding technique to use depends on the specific problem and the nature of the categorical data.

### **13. What is conditional probability? If I call 2 Ubers and 3 Lyfts, what is the probability that all the Lyfts arrive first?**

Conditional probability is the probability of an event occurring given that another event has already occurred. It is calculated by dividing the probability of both events occurring by the probability of the condition.

It is denoted as P(A|B), which reads as "the probability of event A, given that event B has occurred."

The formula for conditional probability is:

P(A|B) = P(A and B) / P(B)

where P(A and B) is the probability of both events A and B occurring together, and P(B) is the probability of event B occurring.

In the given scenario, the probability of a Lyft arriving first is 3/5, since there are 3 Lyfts out of 5 total rides. Assuming the rides arrive independently of each other, the probability of all three Lyfts arriving first is:

(3/5) * (2/4) * (1/3) = 0.1 or 10%

This is calculated by multiplying the probability of the first Lyft arriving first (3/5), the probability of the second Lyft arriving next (2/4, since one Lyft has already arrived), and the probability of the third Lyft arriving last (1/3, since two Lyfts have already arrived). 

Therefore, the probability of all the Lyfts arriving first is 10%.

### **15. What is Naive Bayes Classifier Algorithm. What do you mean by ‘Naive’ in a Naive Bayes?**

The Naive Bayes Classifier is a probabilistic algorithm used for classification tasks in machine learning. It is based on the Bayes theorem and assumes that the presence of a particular feature in a class is independent of the presence of other features.

Bayes' theorem states that the probability of a hypothesis (in this case, the hypothesis that a sample belongs to a particular class) is equal to the prior probability of the hypothesis, multiplied by the likelihood of the observed evidence given the hypothesis, divided by the probability of the observed evidence:

P(hypothesis | evidence) = P(hypothesis) * P(evidence | hypothesis) / P(evidence)

The term 'naive' in Naive Bayes refers to the assumption of independence between the features, which is often not true in real-world scenarios. Despite this simplifying assumption, the Naive Bayes Classifier is widely used and can perform well in many applications, especially in text classification and spam filtering.
